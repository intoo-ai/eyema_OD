```python
import cv2
import os

def parse_yolo_label(label_str):
    # Parse YOLO label string into a list of values
    label_values = list(map(float, label_str.split()))
    class_id, x_center, y_center, width, height = label_values
    return class_id, x_center, y_center, width, height

def create_intermediate_label(label1, label2, t):
    # Interpolate between two labels based on time parameter (t)
    class_id = label1[0]
    x_center = label1[1] + t * (label2[1] - label1[1])
    y_center = label1[2] + t * (label2[2] - label1[2])
    width = label1[3] + t * (label2[3] - label1[3])
    height = label1[4] + t * (label2[4] - label1[4])
    
    return class_id, x_center, y_center, width, height

def follow_objects_between_frames(frame_102_label, frame_140_label, num_frames):
    # Parse YOLO labels for frames 102 and 140
    label_102 = parse_yolo_label(frame_102_label)
    label_140 = parse_yolo_label(frame_140_label)
    
    # Initialize an empty list to store intermediate labels
    intermediate_labels = []

    # Generate intermediate labels for each frame
    for frame in range(103, 140):
        # Calculate the time parameter (t) for interpolation
        t = (frame - 102) / (140 - 102)
        
        # Create intermediate label for the current frame
        intermediate_label = create_intermediate_label(label_102, label_140, t)
        
        # Append the intermediate label to the list
        intermediate_labels.append(intermediate_label)

    return intermediate_labels

def apply_labels_to_frame(frame, label):
    # Extract label values
    class_id, x_center, y_center, width, height = label
    
    # Get image dimensions
    height, width, _ = frame.shape
    
    # Calculate bounding box coordinates
    x_min = int((x_center - width / 2) * width)
    y_min = int((y_center - height / 2) * height)
    x_max = int((x_center + width / 2) * width)
    y_max = int((y_center + height / 2) * height)
    
    # Draw bounding box on the frame
    cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
    
    return frame

def write_labels_to_file(output_file, labels):
    with open(output_file, 'w') as file:
        data = ""
        for _ in labels:
            label = [round(num, 6) for num in _]
            data += f"{' '.join(map(str, label))}\n"
        file.write(data)


# Example usage
first_frame_label = {"obj1":"0 0.343086 0.636333 0.118797 0.149889", 
                     "obj2":"0 0.578000 0.508653 0.102875 0.127861",
                     "obj3":"0 0.670797 0.318514 0.061187 0.079028", 
                     "obj4":"0 0.545102 0.224625 0.122297 0.195194",
                     }

last_frame_label = {"obj1":"0 0.484023 0.894792 0.154297 0.194028",
                    "obj2":"0 0.806695 0.683542 0.138391 0.162639",
                    "obj3":"0 0.555523 0.281514 0.051453 0.069194", 
                    "obj4":"0 0.458961 0.204847 0.102922 0.173028",
                     }

results = {}
for key in first_frame_label.keys():
    results[key] = follow_objects_between_frames(first_frame_label[key], last_frame_label[key], num_frames=39)


label_folder = "/Users/miladsoleymani/Desktop/eyema_OD/dataset/costum_data/labels"

# Print intermediate labels
for idx,_ in enumerate(range(103, 140)):
    labels = []
    for key in results.keys():
        labels.append(results[key][idx])

    write_labels_to_file(os.path.join(label_folder, f"frame_0{_}.txt"), labels)

```

```python
import os
import cv2

def parse_yolo_label(label_str):
    # Parse YOLO label string into a list of values
    label_values = list(map(float, label_str.split()))
    class_id, x_center, y_center, width, height = label_values
    return class_id, x_center, y_center, width, height

def apply_labels_to_image(image_path, label_path, output_folder):
    # Read the image
    image = cv2.imread(image_path)
    
    # Read the YOLO label file
    with open(label_path, 'r') as label_file:
        for line in label_file:
            yolo_label = line.strip()
    
            # Parse YOLO label
            label = parse_yolo_label(yolo_label)
            
            # Get image dimensions
            height, width, _ = image.shape
            
            # Calculate bounding box coordinates
            x_center = int(label[1] * width)
            y_center = int(label[2] * height)
            box_width = int(label[3] * width)
            box_height = int(label[4] * height)
            
            # Calculate top-left and bottom-right coordinates
            x_min = max(0, x_center - box_width // 2)
            y_min = max(0, y_center - box_height // 2)
            x_max = min(width, x_center + box_width // 2)
            y_max = min(height, y_center + box_height // 2)
            
            # Draw bounding box on the image
            cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
            
        # Save the modified image
        output_path = os.path.join(output_folder, os.path.basename(image_path))
        cv2.imwrite(output_path, image)

def apply_labels_to_images(image_folder, label_folder, output_folder):
    # Iterate over images in the specified folder
    for image_name in os.listdir(image_folder):
        if image_name.endswith('.jpg'):
            image_path = os.path.join(image_folder, image_name)
            label_path = os.path.join(label_folder, image_name.replace('.jpg', '.txt'))
            
            # Apply labels to the image
            apply_labels_to_image(image_path, label_path, output_folder)

# Example usage
image_folder = "/Users/miladsoleymani/Desktop/eyema_OD/dataset/costum_data/images"
label_folder = "/Users/miladsoleymani/Desktop/eyema_OD/dataset/costum_data/labels"
output_folder = "/Users/miladsoleymani/Desktop/eyema_OD/dataset/costum_data/results"

apply_labels_to_images(image_folder, label_folder, output_folder)
```

```python
import cv2
import os

def convert_images_to_video(image_folder, output_video_path, fps=25):
    # Get the list of image filenames in the folder
    image_files = sorted([file for file in os.listdir(image_folder) if file.endswith('.jpg')])
    
    # Read the first image to get dimensions
    first_image_path = os.path.join(image_folder, image_files[0])
    first_image = cv2.imread(first_image_path)
    height, width, _ = first_image.shape
    
    # Create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Change codec as needed
    video_writer = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

    # Write each image to the video
    for image_file in image_files:
        image_path = os.path.join(image_folder, image_file)
        frame = cv2.imread(image_path)
        video_writer.write(frame)

    # Release the VideoWriter object
    video_writer.release()

# Example usage
image_folder = "/Users/miladsoleymani/Desktop/eyema_OD/dataset/costum_data/results"
output_video_path = "/Users/miladsoleymani/Desktop/eyema_OD/dataset/costum_data/video.mp4"

convert_images_to_video(image_folder, output_video_path)
```
